# Assignment 2: Titanic Dataset Analysis
## Objective
Build a predictive model for Titanic survival by performing data cleaning, feature engineering, and feature selection.

### Part 1: Data Cleaning
In this section, we handle missing values and outliers to ensure data consistency.

```python
import sys
sys.path.append('../scripts')
from data_cleaning import clean_data
import pandas as pd

df_cleaned = clean_data('../data/train.csv', '../data/train_cleaned.csv')
df_cleaned.head()
```

**Decisions Made:**
- **Age**: Imputed with the median to maintain the distribution without being skewed by outliers.
- **Embarked**: Imputed with the mode ('S') as it had very few missing values.
- **Fare**: Capped outliers using the Interquartile Range (IQR) method to reduce the influence of extreme values.
- **Cabin**: Created a `HasCabin` indicator and filled missing values with 'Unknown' for deck extraction.

### Part 2: Feature Engineering
We derive new insights from the raw data to improve model performance.

```python
from feature_engineering import engineer_features

df_engineered = engineer_features('../data/train_cleaned.csv', '../data/train_engineered.csv')
df_engineered.head()
```

**Features Created:**
- **Title**: Extracted from 'Name' to capture social status.
- **FamilySize**: Combined 'SibSp' and 'Parch'.
- **IsAlone**: Boolean indicator for passengers traveling solo.
- **Deck**: Extracted first letter from 'Cabin'.
- **Fare_Log**: Log transformation to handle skewness in fare distribution.

### Part 3: Feature Selection
We rank features based on their predictive power using a Random Forest Classifier.

```python
from feature_selection import select_features

selected_features = select_features('../data/train_engineered.csv', '../data/selected_features.txt')
print("Selected Features:", selected_features)
```
